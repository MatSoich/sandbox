```python
import pandas as pd
import yfinance as yf

# 株価データを取得する銘柄を指定
tickers = ['AAPL', 'AMZN', 'GOOGL', 'MSFT']
# 株価データを取得してデータフレームに格納
df_list = []

for ticker in tickers:
    stock_data = yf.download(ticker, start='2022-01-01', end='2022-01-10')
    stock_data['Ticker'] = ticker
    df_list.append(stock_data)
df = pd.concat(df_list)

# データフレームをエクセルファイルに書き込む
df.to_excel('stock_data.xlsx', index=False)
```
